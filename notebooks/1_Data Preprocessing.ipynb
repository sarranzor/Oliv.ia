### 4. Transformación de attributes y categories

#### 4.1. Obtención de todos los atributos únicos y sus posibles valores

##### 4.1.1. Limpieza de la variable

```python
import pandas as pd
import ast

def clean_value(value):
    if isinstance(value, str):
        try:
            posible_dict = ast.literal_eval(value)
            if isinstance(posible_dict, dict):
                return clean_dict(posible_dict)
        except (ValueError, SyntaxError):
            pass
    
        if value.startswith("u'") and value.endswith("'"):
            value = value[2:-1]
        elif value.startswith("'") and value.endswith("'"):
            value = value[1:-1]
        
        value = value.lower().capitalize()
    
    return value

def clean_dict(dic):
    clean_dic = {}
    for k, v in dic.items():
        if isinstance(v, dict):
            clean_dic[k] = clean_dict(v)
        else:
            clean_dic[k] = clean_value(v)
    return clean_dic

yelp_business["attributes"] = yelp_business["attributes"].apply(lambda x: clean_dict(x) if isinstance(x, dict) else x)
```

##### 4.1.2. Obtención de los valores únicos

```python
from collections import defaultdict

attributes = defaultdict(set)

def get_values(dic, prefix=''):
    for key, value in dic.items():
        complete_key = f"{prefix}.{key}" if prefix else key
        if isinstance(value, dict):
            get_values(value, prefix=complete_key)
        else:
            attributes[complete_key].add(value)

for dic in yelp_business['attributes']:
    if isinstance(dic, dict):
        get_values(dic)

attributes = dict(attributes)

filename = "attributes.txt"
with open(filename, "w", encoding = "utf-8") as f:
    for key, value in attributes.items():
        f.write(f"{key}: {value}\n")
print("Atributos almacenados en:", filename)
```

#### 4.2. To text

```python
yelp_business_text = yelp_business.copy()
```

##### 4.2.1. Attributes

```python
from utils import attributes_mapping

def attributes_to_text(dic, mapping, prefix=''):
    text = []
    for key, value in dic.items():
        complete_key = f"{prefix}.{key}" if prefix else key
        
        if isinstance(value, dict):
            text.extend(attributes_to_text(value, mapping, prefix=complete_key))
        else:
            if complete_key in mapping:
                mapped_sentence = mapping[complete_key]
                if callable(mapped_sentence):
                    result = mapped_sentence(value)
                    if isinstance(result, str) and result.strip():
                        text.append(result.strip())
                elif isinstance(mapped_sentence, str):
                    val_clean = str(value).strip()
                    if val_clean.lower() not in ['none', 'nan', '']:
                        text.append(f"{mapped_sentence}: {val_clean}")
    
    return text

yelp_business_text["attributes"] = yelp_business["attributes"].apply(
    lambda dic: " | ".join(attributes_to_text(dic, attributes_mapping) if isinstance(dic, dict) else "")
)

yelp_business_text.head()
```

##### 4.2.2. Categories y variable final

```python
def features_to_embed(categories, attributes):
    cats = ", ".join(categories) if isinstance(categories, list) else ""
    return f"Categories: {cats}. Attributes: {attributes.strip()}"

yelp_business_text["features_to_embed"] = yelp_business_text.apply(
    lambda row: features_to_embed(row["categories"], row["attributes"]),
    axis=1
)

yelp_business_text.head()
```

### 4.3. To MultiHot

#### 4.3.1. Attributes

```python
def flatten_attributes(dic, prefix=''):
    items = {}
    for k, v in dic.items():
        key = f"{prefix}.{k}" if prefix else k
        if isinstance(v, dict):
            items.update(flatten_attributes(v, prefix=key))
        else:
            items[key] = v
    return items

def clean_value(val):
    val = str(val).lower().strip()
    if val in ['false', 'none', 'nan', '']:
        return pd.NA
    return val


flat_attributes = yelp_business["attributes"].apply(
    lambda x: flatten_attributes(x) if isinstance(x, dict) else {}
)


flat_attributes = pd.json_normalize(flat_attributes)


flat_attributes = flat_attributes.map(clean_value)

flat_attributes = pd.get_dummies(flat_attributes, dummy_na=False)

flat_attributes.columns = flat_attributes.columns.str.replace('_true$', '', regex=True)

flat_attributes = flat_attributes.map(lambda x: 1 if x else 0)

print(flat_attributes.shape)
flat_attributes.head()
```

#### 4.3.2. Categories

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()

flat_categories = pd.DataFrame(
    mlb.fit_transform(yelp_business["categories"]),
    columns=mlb.classes_,
    index=yelp_business.index
)

flat_categories.columns = [col.replace(" ", "_") for col in flat_categories.columns]

print(flat_categories.shape)
flat_categories.head()
```

#### 4.3.3. DataFrame final 

```python
yelp_business_mh = pd.concat([yelp_business.reset_index(drop=True),
                              flat_attributes.reset_index(drop=True),
                              flat_categories.reset_index(drop=True)],
                             axis=1)

print(yelp_business_mh.shape)
yelp_business_mh.head()
```
